```python
mean_speed = np.loadtxt('mean_speed.txt')

mean_speed = mean_speed.reshape((len(L),len(N),len(Eta),len(V)))

std_speed = np.loadtxt('std_speed.txt')

std_speed = std_speed.reshape((len(L),len(N),len(Eta),len(V)))

print(L)
print(N)
print(Eta)
print(V)
```

##### Plotting the results of a computation

```python
plt.rcParams.update({'font.size': 10})

plt.figure(figsize=(8,8))
for i in range(len(L)):
    for j in range (len(N)):
        for l in  range (len(V)):
            plt.errorbar(Eta, mean_speed[i,j,:,l], yerr=std_speed[i,j,:,l], \
                             label="L = " + str(L[i]) + ", N = " + str(N[j]) + \
                                     ", V = " + str(V[l]))
plt.xlabel('$\eta$')
plt.ylabel('$v_a$')
plt.legend()
plt.show()
```
